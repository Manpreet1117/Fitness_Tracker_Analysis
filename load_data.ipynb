```python
import pandas as pd
from sqlalchemy import create_engine

# ⚠️ Replace 'yourpassword' with your actual PostgreSQL password
engine = create_engine('postgresql://postgres:171223@localhost:5432/fitness_tracker')

# Load CSVs
daily_activity = pd.read_csv('data/dailyActivity_merged.csv')
sleep_day = pd.read_csv('data/sleepDay_merged.csv')
hourly_steps = pd.read_csv('data/hourlySteps_merged.csv')

# Check shapes
print("Daily Activity:", daily_activity.shape)
print("Sleep Day:", sleep_day.shape)
print("Hourly Steps:", hourly_steps.shape)

# Push to PostgreSQL
daily_activity.to_sql('daily_activity', engine, if_exists='replace', index=False)
sleep_day.to_sql('sleep_day', engine, if_exists='replace', index=False)
hourly_steps.to_sql('hourly_steps', engine, if_exists='replace', index=False)

print("✅ All tables loaded successfully!")
```
